```python
from pyspark.sql.functions import (
    col,
    avg,
    min,
    max,
    count,
    round
)

# Read silver Selic table
df = spark.table("portfolio.bcb_economic_indicators.silver_selic_daily")

# Create monthly analytical table
monthly_df = (
    df
    .groupBy("year", "month")
    .agg(
        count("*").alias("total_records"),
        round(avg("selic_rate"), 4).alias("avg_selic_rate"),
        round(min("selic_rate"), 4).alias("min_selic_rate"),
        round(max("selic_rate"), 4).alias("max_selic_rate")
    )
    .orderBy("year", "month")
)

# Save as gold Delta table
monthly_df.write.mode("overwrite").saveAsTable(
    "portfolio.bcb_economic_indicators.gold_monthly_selic"
)

display(monthly_df)
```
