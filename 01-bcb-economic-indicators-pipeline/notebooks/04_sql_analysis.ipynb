```python
%sql
SELECT *
FROM portfolio.bcb_economic_indicators.gold_monthly_selic
ORDER BY year, month;
```

```python
%sql
SELECT
  year,
  ROUND(AVG(avg_selic_rate), 4) AS annual_avg_selic_rate,
  MIN(min_selic_rate) AS annual_min_selic_rate,
  MAX(max_selic_rate) AS annual_max_selic_rate
FROM portfolio.bcb_economic_indicators.gold_monthly_selic
GROUP BY year
ORDER BY year;
```

```python
%sql
SELECT
  year,
  month,
  avg_selic_rate
FROM portfolio.bcb_economic_indicators.gold_monthly_selic
ORDER BY avg_selic_rate DESC
LIMIT 5;
```

```python
%sql
SELECT
  year,
  month,
  avg_selic_rate
FROM portfolio.bcb_economic_indicators.gold_monthly_selic
ORDER BY avg_selic_rate ASC
LIMIT 5;
```

```python
%sql
SELECT
  CONCAT(CAST(year AS STRING), '-', LPAD(CAST(month AS STRING), 2, '0')) AS year_month,
  avg_selic_rate,
  min_selic_rate,
  max_selic_rate
FROM portfolio.bcb_economic_indicators.gold_monthly_selic
ORDER BY year, month;
```
